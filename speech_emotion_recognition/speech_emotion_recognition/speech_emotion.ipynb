# Input speech signal

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from librosa import display

data, sampling_rate = librosa.load('recorded.wav')
print('data:', data.shape, '\nsample_rate:', sampling_rate)
```

# Feature Extraction

### Log-Mel Spectrogram

```python
# Feature extraction：Log-Mel Spectrogram
melspec = librosa.feature.melspectrogram(data, sampling_rate, n_fft=1024, hop_length=512, n_mels=128)

print('Mel frequency domain:',melspec.shape[0],'\nMel time domain:',melspec.shape[1],'\nshow data:\n',melspec)

logmelspec = librosa.power_to_db(melspec)
print('Mel frequency domain:',logmelspec.shape[0],'\nMel time domain:',logmelspec.shape[1],'\nshow data:\n',logmelspec)
```

### MFCC

```python
mfccs = librosa.feature.mfcc(y = data, sr = sampling_rate, n_mfcc = 40)
print('mfccs\n', mfccs.shape, mfccs)
```

```python
plt.figure(figsize = (12, 4))
librosa.display.waveplot(data, sampling_rate)
```

```python
# spectrogram
plt.figure(figsize = (12, 4))
librosa.display.specshow(logmelspec, sr=sampling_rate, x_axis='time', y_axis='mel')
plt.title('Mel spectrogram')
plt.show()
```
